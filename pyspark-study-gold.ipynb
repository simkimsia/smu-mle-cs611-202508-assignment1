```python
import pyspark
from pyspark.sql import SparkSession

# Initialize Spark
spark = SparkSession.builder.appName("parquet_explorer").master("local[*]").getOrCreate()
spark.sparkContext.setLogLevel("ERROR")

# Example: Load and explore feature store
df = spark.read.parquet('datamart/20250929_074849/gold/feature_store/gold_feature_store_2023_01_01.parquet')

print("Schema:")
df.printSchema()

print(f"\nRow count: {df.count()}")

print("\nFirst 10 rows:")
df.show(10)

print("\nColumn names:")
print(df.columns)

spark.stop()
```
